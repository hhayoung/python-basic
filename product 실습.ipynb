```python
class Product:
    
    def __init__(self, category, p_code, p_name, price, description):
        self.category = category
        self.p_code = p_code
        self.p_name = p_name
        self.price = price
        self.description = description

    def _info(self):
        print('카테고리:', self.category )
        print('제품 코드:', self.p_code)
        print('제품 이름:', self.p_name)
        print('제품 가격:', self.price)
        print('제품 설명:', self.description)

# 화면 출력시 제품 정보 입력 메소드
def set_info():
    
    category = input('카테고리:')
    p_code = input('제품 코드:')
    p_name = input('제품 이름:')
    price = input('제품 가격:')
    description = input('제품 설명:')
    
    product = Product(category, p_code, p_name, price, description)
    return product

def print_product_info(product_list):
    print('-'*20)
    for product in product_list:
        product._info()
        print('-'*20)

# 메뉴 화면 
def menu_screen():
    print('='*20)
    print('1.제품 정보 입력')
    print('2.제품 정보 출력')
    print('3.제품 정보 삭제')
    print('4.종료 ')
    print('='*20)
    
    menu = input('메뉴를 선택하시오:')
    return int(menu)

# 메뉴 중 3번에 해당하는 삭제 함수
def delete_pro(product_list,p_code):
    for index, product in enumerate(product_list):
        if product.p_code == p_code:
            del product_list[index]            
        
def save_product(product_list):
    with open('./product_db.txt', 'w',encoding='utf8') as f:
        for product in product_list:
            f.write(product. + '\n')
            f.write(product. + '\n')
            f.write(product. + '\n')
            f.write(product. + '\n')
            
# product_list 작성할 수 있게 하는 반복문
def run():
    
    product_list = []
    
    while True:
        menu = menu_screen()
        if menu == 1: 
            product = set_info()
            product_list.append(product)
        elif menu == 2: 
            print_product_info(product_list)
        elif menu == 3:
            p_code = input('삭제할 코드명을 입력하세요')
            delete_pro(product_list,p_code)
        elif menu == 4: 
            print('종료합니다')
            break
        else: 
            print('다시 입력하세요')
            
            
            
if __name__=='__main__':
    run()
```
